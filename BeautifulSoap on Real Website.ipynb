```python
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

```python
response = requests.get('https://www.imdb.com/search/title/?release_date=2018&sort=num_votes,desc&page=1&ref_=adv_nxt')
data = BeautifulSoup(response.text, 'html.parser')
data
```

### From this link,
### Find and print the name and genre of the first 3 titles

```python
h3 = data.find_all('h3')
p = data.find_all(class_ = 'genre')
title = []
genre = []
for i in range(0, 3):
    title.append(h3[i].a.string)
    genre.append(p[i].string.strip().split(', '))
for i in range(0, len(title)):
    print(title[i], end = " ; ")
    for j in range(0, len(genre[i])):
        if (j == len(genre[i]) - 1):
            print(genre[i][j])
            continue
        print(genre[i][j], end = ", ")
```
